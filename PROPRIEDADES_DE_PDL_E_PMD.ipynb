# Perda Dependente de Polarização (PDL), Dispersão por Modos de Polarização (PMD) e Ganho Dependente de Polarização (PDG)

1.  Perda Dependente de Polarização (PDL):
A PDL refere-se à perda de energia que ocorre de forma preferencial dependendo do estado de polarização da luz. Em sistemas ópticos, essa perda pode afetar a intensidade do sinal dependendo de como a luz é polarizada. No modelo de Jones, um dos eixos de polarização sofre mais perda que o outro, o que altera o estado de polarização de saída. A PDL é independente do comprimento de onda da luz.

2. Dispersão por Modos de Polarização (PMD):
A PMD é um fenômeno de atraso temporal entre diferentes modos de polarização, causado por segmentos birefringentes concatenados. Cada segmento homogêneo de fibra óptica gera um atraso de grupo diferencial entre os modos. A PMD ocorre quando esses segmentos são colocados em cascata, produzindo uma distorção temporal do sinal. A PMD é dependente do comprimento de onda e tende a despolarizar o sinal devido à dispersão do estado de polarização ao longo da frequência.

3. Ganho Dependente de Polarização (PDG):
O PDG é o efeito oposto da PDL e está relacionado ao ganho preferencial de certos estados de polarização em amplificadores ópticos. Em amplificadores, alguns estados de polarização podem ter um ganho maior que outros, o que pode resultar em um aumento de ruído e distorção do sinal, especialmente em altas velocidades de transmissão. Assim como a PDL, o PDG pode afetar negativamente a comunicação óptica.

4. Interação entre PDL e PMD:
Quando a PDL e a PMD se combinam, os efeitos podem ser mais complexos e impactar mais significativamente o sistema de comunicação. A PDL é geralmente independente do comprimento de onda, enquanto a PMD é dependente do comprimento de onda. A combinação de ambas pode resultar em uma PDL dependente de comprimento de onda. Além disso, a PDL pode afetar o atraso de grupo diferencial da PMD.

5. Impacto na Polarização Parcial:
A PDL tende a polarizar um sinal parcialmente polarizado, pois atua como um polarizador parcial. Por outro lado, a PMD tende a despolarizar o sinal, criando dispersão no estado de polarização ao longo da frequência. Dispositivos como despolarizadores de Lyot exploram essa dispersão para despolarizar o sinal, mas a introdução de PMD no sistema pode complicar a reversibilidade do processo.

6. Matriz de Mueller:
A matriz de Mueller é usada para representar os efeitos de PDL e PMD no estado de polarização. A PDL é representada por uma matriz de Jones hermitiana e, quando convertida para a matriz de Mueller, ocupa todos os 16 espaços possíveis. Já a PMD é representada por uma matriz de Jones traceless, o que permite que seu efeito seja reversível em princípio. A cascata de birefringentes com PDL espalha os efeitos da PMD nos 16 espaços da matriz de Mueller, tornando a combinação irreversível.

**Perda Dependente de Polarização (PDL)**

A PDL pode surgir de vários fatores ao longo de um link de fibra óptica, como componentes micro-ópticos, filtros integrados, acopladores de fibra fundida, amplificadores ópticos, micro-dobramentos na fibra e materiais como guias de onda de polímero. Esses efeitos ocorrem quando há acoplamento desigual de diferentes polarizações ou devido às propriedades dos materiais, sendo todos descritos por um único modelo.

**Definição de PDL e Cálculo em Decibéis**

A perda dependente de polarizaçã(𝜌𝑑𝐵) é uma medida da diferença entre as intensidades máximas e mínimas de transmissão de um sinal em função de sua polarização. Ela é expressa em decibéis e definida pela seguinte equação:

Onde:

* 𝑇𝑚𝑎𝑥 é a intensidade de transmissão máxima (ocorre quando o vetor de polarização da luz está alinhado com o eixo de transmissão máximo do elemento PDL), sempre será maior que 𝑇𝑚𝑖𝑛.

* 𝑇𝑚𝑖𝑛 é a intensidade de transmissão mínima (ocorre para a polarização ortogonal, mesmo na presença de PMD).

* O valor de 𝑇𝑚𝑎𝑥 é sempre maior ou igual a 𝑇𝑚𝑖𝑛 porque 𝑇𝑚𝑎𝑥 ocorre quando a luz está alinhada com o eixo de máxima transmissão (sem perda), enquanto 𝑇𝑚𝑖𝑛 ocorre quando a luz está alinhada com o eixo ortogonal (máxima perda). Isso resulta em uma PDL positiva ou zero.

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando os dados
alpha = np.linspace(0, 3, 100)  # Coeficiente de perda
Tmax = 1 / (1 + np.tanh(alpha))  # Intensidade máxima
Tmin = 1 / (1 - np.tanh(alpha))  # Intensidade mínima

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(alpha, Tmax, label=r'$T_{max}$', color='b', linewidth=2)
plt.plot(alpha, Tmin, label=r'$T_{min}$', color='r', linestyle='--', linewidth=2)
plt.fill_between(alpha, Tmax, Tmin, color='gray', alpha=0.3, label=r'$T_{max} \geq T_{min}$')

# Adicionando títulos e legendas
plt.title('Gráfico de Intensidade Máxima (Tmax) e Mínima (Tmin) em Função do Coeficiente de Perda', fontsize=14)
plt.xlabel('Coeficiente de Perda (α)', fontsize=12)
plt.ylabel('Intensidade de Transmissão (T)', fontsize=12)
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

O código a seguir tem como objetivo calcular a intensidade de transmissão de um sinal após passar por um elemento com Perda Dependente de Polarização (PDL). Ele faz isso utilizando a matriz de Jones, que é uma ferramenta matemática usada para descrever a propagação da luz polarizada através de meios ópticos Aqui, o coeficiente de perda
𝛼 é definido como 0.345, que corresponde a uma perda de 3 dB (decibéis). O vetor de polarização de entrada s é definido como [1, 0], que representa uma luz polarizada horizontalmente (ou seja, a luz está completamente na direção
𝑥).

A função transmission_intensity(alpha,s) é chamada para calcular a intensidade de transmissão. O valor resultante é impresso no final.

.

```python
import numpy as np

# Função para calcular a matriz de Jones para PDL
def jones_matrix(alpha):
    return np.array([[1, 0], [0, np.exp(-alpha)]])

# Função para calcular a intensidade de transmissão
def transmission_intensity(alpha, s):
    # Normalizando o vetor de polarização de entrada
    s_norm = s / np.linalg.norm(s)
    # Matriz de Jones
    PDL_matrix = jones_matrix(alpha)
    # Calculando a intensidade transmitida
    transmission = np.dot(s_norm.T, np.dot(PDL_matrix.T, np.dot(PDL_matrix, s_norm)))
    return transmission

# Definindo o coeficiente de perda
alpha = 0.345  # Para PDL de 3 dB

# Vetor de polarização de entrada
s = np.array([1, 0])  # Polarização horizontal

# Calculando a transmissão
T_p = transmission_intensity(alpha, s)
print(f'Intensidade de transmissão: {T_p}')
```

* Matriz de Jones (PDL) -- Função jones_matrix(alpha) transforma o vetor de polarização de entrada s.

* PDL matrix𝑇 -- Função transmission_intensity(alpha, s)
  é a transposta da matriz de Jones (porque estamos lidando com vetores coluna e queremos a forma correta de multiplicação).
* O produto escalar da polarização de entrada e saída é o que dá a intensidade transmitida.

**Mudança de Estado de Polarização em Elementos PDL**

Em um elemento com Perda Dependente de Polarização (PDL), o estado de polarização da luz de saída geralmente difere do estado de polarização da luz de entrada. Este fenômeno pode ser observado de maneira simples: quando um sinal circular direito (polarização circular direita) é transmitido por um elemento PDL, um dos eixos da polarização é reduzido, e, portanto, a polarização do sinal muda de circular para elíptica. Esse tipo de alteração ocorre porque a PDL afeta as componentes do sinal de forma dependente de sua polarização, resultando em uma distorção do estado de polarização.
